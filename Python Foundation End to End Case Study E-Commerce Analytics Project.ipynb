### b. Understanding how many new customers acquired every month

```python
customer_order = pd.merge(CUSTOMERS, ORDERS, on = 'customer_id')
```

```python
customer_order.order_purchase_timestamp = pd.to_datetime(customer_order.order_purchase_timestamp)
```

```python
first_order = customer_order.groupby('customer_id')['order_purchase_timestamp'].min().reset_index()
```

```python
first_order['first_order_month'] = first_order['order_purchase_timestamp'].dt.to_period('M')
```

```python
first_order.rename(columns = {'order_purchase_timestamp': 'first_order_date'}, inplace = True)
```

```python
customer_order['order_month'] = customer_order.order_purchase_timestamp.dt.to_period('M')
```

```python
merged_data = pd.merge(customer_order, first_order, on = 'customer_id')
```

```python
merged_data['customer_type'] = merged_data.apply(lambda x : 'new_customer' if x['order_month'] == x['first_order_month'] else 'repeat_customer' , axis = 1 )
```

```python
result = merged_data.groupby(['order_month', 'customer_type'])['customer_id'].count().reset_index()
```

```python
result
```

### e. Understand the trends/seasonality of sales, quantity by category, location, month, week, day, time, channel, payment method etcâ€¦â€¦

```python
customer = pd.merge(ORDER_ITEMS, PRODUCTS, on = 'product_id')
df = pd.merge(customer, ORDERS, on = 'order_id')
df1 = pd.merge(df, CUSTOMERS, on = 'customer_id')
final_customer = pd.merge(df1,ORDER_PAYMENTS , on = 'order_id')
```

```python
# sales, quantity by category

sales_qty = customer.groupby('product_category_name')[['price', 'product_photos_qty']].sum().reset_index()
sales_qty.rename(columns = {'product_category_name': 'Categories' ,'price' : 'Sales','product_photos_qty' : 'QTY'}, inplace = True)
sales_qty
```

```python
# sales, quantity by location

location = final_customer.groupby('customer_state')[['price', 'product_photos_qty']].sum().reset_index()
location.rename(columns = {'customer_state': 'Location' ,'price' : 'Sales','product_photos_qty' : 'QTY'}, inplace = True)
location
```

```python
# sales and QTY by month
final_customer['order_purchase_timestamp'] = pd.to_datetime(final_customer['order_purchase_timestamp'])
month = final_customer.groupby(final_customer['order_purchase_timestamp'].dt.to_period('M'))[['price', 'product_photos_qty']].sum().reset_index()
month.rename(columns = {'order_purchase_timestamp': 'Month' ,'price' : 'Sales','product_photos_qty' : 'QTY'}, inplace = True)
month
```

```python
# sales and QTY by week
week = final_customer.groupby([final_customer['order_purchase_timestamp'].dt.to_period('M'),final_customer['order_purchase_timestamp'].dt.isocalendar().week])[['price', 'product_photos_qty']].sum().reset_index()
week.rename(columns = {'order_purchase_timestamp': 'Month' ,'price' : 'Sales','product_photos_qty' : 'QTY'}, inplace = True)
week.sort_values(by = 'Month')
```

```python
# sales and QTY by day

day = final_customer.groupby([final_customer['order_purchase_timestamp'].dt.day_name()])[['price', 'product_photos_qty']].sum().reset_index()
day.rename(columns = {'order_purchase_timestamp': 'Day_Name' ,'price' : 'Sales','product_photos_qty' : 'QTY'}, inplace = True)
day
```

```python
# sales and QTY by time
time = final_customer.groupby([final_customer['order_purchase_timestamp'].dt.strftime('%I %p')])[['price', 'product_photos_qty']].sum().reset_index()
time.rename(columns = {'order_purchase_timestamp': 'Time' ,'price' : 'Sales','product_photos_qty' : 'QTY'}, inplace = True)
time
```

```python
# sales and QTY by payment method

payment_method = final_customer.groupby([final_customer['payment_type']])[['price', 'product_photos_qty']].sum().reset_index()
payment_method.rename(columns = {'payment_type': 'Payment Method' ,'price' : 'Sales','product_photos_qty' : 'QTY'}, inplace = True)
payment_method
```

### 2. Performing Customers/sellers Segmentationd 

#### a. Divide the customers into groups based on the revenue generated

```python
customer = pd.merge( ORDER_ITEMS, ORDERS , on = ['order_id'])

revenue = pd.pivot_table(customer, index = 'customer_id', values = 'price', aggfunc = 'sum').reset_index()

revenue['customer_type'] = pd.cut(revenue.price, bins = [0,500,2500,5000,14000] ,labels = ['New', 'Repeat', 'Prime', 'Elite'])

revenue
```

#### b. Divide the sellers into groups based on the revenue generated 

```python
seller = ORDER_ITEMS.groupby('seller_id')['price'].sum().reset_index()
seller['seller_type'] = pd.cut(seller.price, bins = [0,50000, 100000, 200000, 250000], labels = ['Silver', 'Gold', 'Paltinum', 'Diamond'])
seller
```

#### c. Average rating by location, seller, product, category, month etc.

```python
ak['review_creation_date'] = pd.to_datetime(ak['review_creation_date'])

seller = ak.groupby('seller_id')['review_score'].mean()
pr = ak.groupby('product_id')['review_score'].mean()
cat = ak.groupby('product_category_name')['review_score'].mean()
rev = ak.groupby(ak['review_creation_date'].dt.month)['review_score'].mean()

print(f' avg rating for {seller}')
print(f' avg rating for {pr}')
print(f' avg rating for {cat}')
print(f' avg rating for {rev}')
```
